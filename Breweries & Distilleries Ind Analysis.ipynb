```python
import datetime as dt
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objects as go
import yfinance as yf

from pandas_datareader import  data as pdf
from plotly.subplots import make_subplots

pyo.init_notebook_mode(connected=True)
pd.options.plotting.backend = 'plotly'
```

```python
stocklist = ['^CRSLDX', 'UNITDSPR.NS', 'UBL.NS', 'RADICO.NS']
stocklist
```

```python
df = yf.download(stocklist, start='2015-01-01', end='2025-01-31').dropna()
df
```

```python
df.Close
```

```python
log_returns = np.log(df.Close/df.Close.shift(1)).dropna()
log_returns
```

```python
daily_std = log_returns.std()
daily_std
```

```python
annualized_valatility = daily_std * np.sqrt(252)
annualized_valatility*100
```

```python
fig = make_subplots(rows=2, cols=2)

image0 = go.Histogram(x=log_returns['^CRSLDX'], name='NIFTY500')
image1 = go.Histogram(x=log_returns['RADICO.NS'], name='RADICO KHAITAN')
image2 = go.Histogram(x=log_returns['UNITDSPR.NS'], name='UNITED SPIRITS')
image3 = go.Histogram(x=log_returns['UBL.NS'], name='UNITED BREWERIES')

fig.append_trace(image0, 1, 1)
fig.append_trace(image1, 1, 2)
fig.append_trace(image2, 2, 1)
fig.append_trace(image3, 2, 2)

fig.update_layout(autosize=False, width=1100, height=600, title='Frequency of log returns')

fig.show()
```

```python
fig = make_subplots(rows=2, cols=2)

image0 = go.Box(x=log_returns['^CRSLDX'], name='NIFTY500')
image1 = go.Box(x=log_returns['RADICO.NS'], name='RADICO KHAITAN')
image2 = go.Box(x=log_returns['UNITDSPR.NS'], name='UNITED SPIRITS')
image3 = go.Box(x=log_returns['UBL.NS'], name='UNITED BREWERIES')

fig.append_trace(image0, 1, 1)
fig.append_trace(image1, 1, 2)
fig.append_trace(image2, 2, 1)
fig.append_trace(image3, 2, 2)

fig.update_layout(autosize=False, width=1100, height=600, title='Frequency of log returns')

fig.show()
```

# Calculating Volatility

```python
TRADING_DAYS = 20

volatility = log_returns.rolling(window=TRADING_DAYS).std() * np.sqrt(TRADING_DAYS)

volatility.plot()
```

# Calculating Sharpe Ratio

```python
Rf = 0.06/252

sharpe_ratio = (log_returns.rolling(window=TRADING_DAYS).mean() - Rf) * TRADING_DAYS/volatility

sharpe_ratio.plot()
```

```python
sortino_volatility = log_returns[log_returns<0].rolling(window=TRADING_DAYS, center=True, min_periods=10).std() * np.sqrt(TRADING_DAYS)

sortino_ratio = (log_returns.rolling(window=TRADING_DAYS).mean() - Rf) * TRADING_DAYS / sortino_volatility

sortino_volatility.plot()
```
