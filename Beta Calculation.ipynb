```python
import datetime as dt
import pandas as pd
import numpy as np
import yfinance as yf

from scipy import stats
from pandas_datareader import data as pdr
```

```python
stocklist = ['^NSEBANK', 'HDFCBANK.NS', 'ICICIBANK.NS', 'AXISBANK.NS','SBIN.NS']
stocklist
```

```python
df = yf.download(stocklist, start='2015-01-01', end='2025-12-31')
df
```

```python
df.Close
```

# Log Returns

```python
log_returns = np.log(df.Close / df.Close.shift(1)).dropna()
log_returns.head()
```

# Beta Calculation

```python
def calc_beta(df):
    market = df.iloc[:, -1]
    stocks = df.iloc[:, :-1]

    betas = {}

    for col in stocks.columns:
        cov = np.cov(stocks[col], market)
        betas[col] = cov[0, 1] / cov[1, 1]

    return pd.Series(betas, name="Beta")

calc_beta(log_returns)
```
